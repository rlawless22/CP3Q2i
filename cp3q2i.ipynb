```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation
def equation(t, y):
    return t**2 + y

# Create a grid of points for direction vectors
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)

X, Y = np.meshgrid(x, y)

# Direction of vectors at each point
U = 1
V = equation(X, Y)

# Normalize direction vectors 
N = np.sqrt(U**2 + V**2)
U2, V2 = U/N, V/N

# Plot the direction field
plt.quiver(X, Y, U2, V2, color='blue')

# Initial conditions
initial_conditions = [(-2, -2, 'red'), (-2, -2.1, 'green'), (-2, -1.9, 'purple')]


t_span = np.linspace(-2, 2, 100)

# Plot solutions
for ic in initial_conditions:
    t0, y0, color = ic
    sol = solve_ivp(equation, [t0, 2], [y0], t_eval=t_span)
    plt.plot(sol.t, sol.y.flatten(), label=f'y({t0}) = {y0}', color=color)

# Plot settings
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Directional field with solution curves')
plt.grid(True)
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.show()
```
